# Testando o modelo

```python
from tensorflow import keras
model = keras.models.load_model('model_reg.keras')
```

```python
img = cv2.imread(current_dir + '\\Testing'  + '\\meningioma_tumor' + '\\image(4).jpg')
img = cv2.resize(img,(150,150))
img_array = np.array(img)
img_array.shape
```

```python
img_array = img_array.reshape(1,150,150,3)
img_array.shape
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(current_dir + '\\Testing'  + '\\meningioma_tumor' + '\\image(4).jpg')
plt.imshow(img,interpolation='nearest')
plt.show()
```

```python
pred = model.predict(img_array)
index = pred.argmax()

if index == 0:
    print(f'indice {index} = glioma tumor')
elif index == 1:
    print(f'indice {index} = meningioma tumor')
elif index == 2:
    print(f'indice {index} = no tumor')
else:
    print(f'indice {index} = pituitary tumor')
```
